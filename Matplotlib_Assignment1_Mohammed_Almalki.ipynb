```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Q1: Create two charts using list and dataframe.

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 20)
rate =[2.5,2.3,3.8,4.5,4.8,4.1]
bins=[1,2,3,4,5]
plt.subplot(1,2,1)
plt.hist(rate , bins=bins);
data= {"Restaurants":['KFC',"McDonald's","Five Guys","8Portionspizza","Blanka","mama noura"] 
      ,"Rates":rate}
#print(data)
res=pd.DataFrame(data)
plt.subplot(1,2,2)

plt.hist(res);
plt.title("reta" ,fontsize=20)
plt.xlabel("Restaurants")
plt.ylabel("Rates")
```

## Q2: Create pie chart with equal explode values.

```python
Restaurants=['KFC',"McDonald's","Five Guys","8Portionspizza","Blanka","mama noura"] 
rate =[2.5,2.3,3.8,4.5,4.8,4.1]

plt.pie(rate,labels=Restaurants,autopct='%1.2f%%',startangle=90,shadow=True,explode=(0,0,0,0,0,0));
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(res['Restaurants'],res['Rates'],marker='*',s=50,alpha=0.4);
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(res['Restaurants'],res['Rates'],marker='*',s=50,alpha=0.4,color='r');
plt.title('Reat of a Restaurants',loc='left')
plt.xlabel('Reat')
plt.ylabel('Restaurants')
```
