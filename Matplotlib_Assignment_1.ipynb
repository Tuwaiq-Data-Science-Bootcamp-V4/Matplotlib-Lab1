```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
gender = ['winner', 'Loser']
values = [45,25]

plt.figure(figsize=(8, 5))
plt.pie(values, labels= gender,autopct='%1.2f%%', colors=['#6cb3ff','#99aa99'], radius=1, explode = [0.1,0])
#plt.xlabel('result')
#plt.ylabel('Values')
plt.title('result')
plt.show()

data = {'X': [100, 200, 300, 400, 500],
        'Y': [205, 450, 215, 300, 100]}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
plt.scatter(df['X'], df['Y'], color='#8c8bc8')
plt.xlabel('X axis ')
plt.ylabel('Yaxis')
plt.title('Scatter Plot ')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
L = ['Python','Java','C++','C#','Flutter']
N = [ 315, 245, 120, 100, 200]

# add colors
#colors = ['#ffc0cb','#f5f5dc','#fffcb','#f5f5dc', '#ffcc39']
plt.pie(N, labels= L, autopct='%1.2f%%' ,colors=['#e6e6fa','#ff9999','#f5f5dc','#f0ffff','#f0e68c'], startangle=180, radius=1.5, explode = [0,0,0,0,0]);

plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x= [100, 200, 300, 400, 500]
y= [205, 450, 215, 300, 100]
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='#1f77b4',marker = 'o',alpha =0.4, s= 50)
plt.xlabel('X axis ', fontsize=50)
plt.ylabel('Y axis', fontsize=50)
plt.title('Scatter Plot ', fontsize=50, color='#1f77b4')
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python


data = {'X': [100, 200, 300, 400, 500],
        'Y': [205, 450, 215, 300, 100]}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 5))
plt.scatter(df['X'], df['Y'], color='#1f77b4')
plt.xlabel('X axis ', fontsize=16)
plt.ylabel('Y axis', fontsize=16)
plt.title('Scatter Plot ', fontsize=16, color='red', loc='left')
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
plt.plot([100, 200, 300, 400, 500], color ='#1f77b4')
plt.plot([205, 400, 215, 300, 100], '#add8e6')
plt.plot([0, 20, 35, 40 , 99], color ='#7bc8f6');
plt.title('Line Charts')
plt.legend(['winner','Loser','Balanced'], shadow = True, loc = 'upper left');
plt.show()
```
