```python
import numpy as np
from numpy import random
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
# Using list
lis = [30,10,40,60,75,80]
plt.hist(lis, facecolor ='#8c8bc8',bins = 8);
```

```python
df = pd.DataFrame({'Recipe':['Spaghetti Bolognese','Chicken Alfredo'
,'Beef Stir-Fry','Vegetable Curry','Margherita Pizza','Chocolate Chip Cookies'],
                  'Rate':[7.5,2.5,10.0,4.5,9.0,8.5]})
```

```python
# Using DataFrame
plt.hist(df['Rate'], facecolor ='#cfcefc',bins = 8);
```

## Q2: Create pie chart with equal explode values.

```python
color = ['#afaefa','#bfbefb','#cfcefc','#dfdffd','#efeffe','#8c8bc8']
plt.pie(df['Rate'], labels=df['Recipe'],autopct='%1.1f%%',colors=color, radius= 1.5);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x =random.randint(100, size=(800))
y =range(0,800)
plt.scatter(x, y, alpha=0.4, s= 50, c= '#8c8bc8');
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(x, y, alpha=0.4, s= 50, c= '#8c8bc8');
plt.title('Scatter Chart',color='r')
plt.xlabel('X axis', fontsize= 16);
plt.ylabel('Y axis', fontsize= 16);
```

## Q5: Create three line charts with Legends at left and colors.

```python
plt.plot(range(100,400,50),[20,300,500,100,700,100], color='#efeffe',linestyle = 'solid', linewidth = 2);
plt.plot(range(100,400,50),[200,30,550,600,650,700], color='#464564',linestyle = 'solid', linewidth = 2);
plt.plot(range(100,400,50),[700,300,400,500,550,100], color='#8c8bc8',linestyle = 'solid', linewidth = 2);
plt.xlim(100,350)
plt.legend(['Line1','Line2','Line3',],shadow =True, loc = 2);
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=[10,15])
plt.suptitle('Subplots',fontsize = 20)

#first chart

plt.subplot(3,1,1)
plt.pie(df['Rate'], labels=df['Recipe'],autopct='%1.1f%%',colors=color, radius= 1.5);

#second chart
plt.subplot(3,1,2)
plt.plot(range(100,400,50),[20,300,500,100,700,100], color='#efeffe',linestyle = 'solid', linewidth = 2);
plt.plot(range(100,400,50),[200,30,550,600,650,700], color='#464564',linestyle = 'solid', linewidth = 2);
plt.plot(range(100,400,50),[700,300,400,500,550,100], color='#8c8bc8',linestyle = 'solid', linewidth = 2);
plt.xlim(100,350);

#third chart
plt.subplot(3,1,3)
plt.hist(df['Rate'], facecolor ='#cfcefc',bins = 8);
```
