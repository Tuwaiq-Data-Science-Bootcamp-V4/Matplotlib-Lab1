```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
Region=['Riyadh','Eastern Region','Makkah Region','Qassim']
Population=[8.6 ,5.1 ,8.0,1.3]
plt.bar(Region,Population)
plt.title('Population by Saudi Region')
plt.xlabel('Region')
plt.ylabel('Population in millions')
plt.show()
```

```python
LifeExpectancy={
    'Year':[2019,2020,2021,2022],
    'Male_age':[76.0,76.1,73.3,75.3]
}
Life_Expectancy=pd.DataFrame(LifeExpectancy)
plt.plot(Life_Expectancy['Year'],Life_Expectancy['Male_age'])
plt.title('Saudi Male Life Expectancy')
plt.xlabel('Years')
plt.ylabel('middle age')
plt.xticks(Life_Expectancy['Year'])
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
explodes = [0.1, 0.1, 0.1, 0.1] 
plt.pie(Population ,labels=Region ,autopct='%1.1f%%',shadow=True ,explode=explodes)
plt.title('Population by Saudi Region')
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(Life_Expectancy['Year'],Life_Expectancy['Male_age'],marker='o', alpha=0.4, s=50)
plt.xticks(Life_Expectancy['Year'])
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(Life_Expectancy['Year'],Life_Expectancy['Male_age'],marker='o', alpha=0.4, s=50)
plt.xticks(Life_Expectancy['Year'])
plt.title('Saudi Male Life Expectancy', color='red', loc='left')
plt.xlabel('Years' ,fontsize = 16)
plt.ylabel('middle age',fontsize = 16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
x = [1, 2, 3, 4, 5]
y1 = [5, 10, 8, 12, 15]
y2 = [3, 6, 50, 8, 10]
y3 = [7, 14, 11, 16, 70]

plt.plot(x, y1, label='Line 1', color='blue')
plt.plot(x, y2, label='Line 2', color='green')
plt.plot(x, y3, label='Line 3', color='red')

plt.title('Three Line Charts with Legends')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.legend(loc='upper left')
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 20)
plt.subplot(1,3,1)
plt.bar(Region,Population)
plt.title('Population by Saudi Region')
plt.xlabel('Region')
plt.ylabel('Population in millions')

plt.subplot(1,3,2)
plt.plot(Life_Expectancy['Year'],Life_Expectancy['Male_age'])
plt.title('Saudi Male Life Expectancy')
plt.xlabel('Years')
plt.ylabel('middle age')
plt.xticks(Life_Expectancy['Year'])

plt.subplot(1,3,3)
plt.scatter(Life_Expectancy['Year'],Life_Expectancy['Male_age'],marker='o', alpha=0.4, s=50)
plt.xticks(Life_Expectancy['Year'])
plt.title('Saudi Male Life Expectancy')
plt.xlabel('Years' )
plt.ylabel('middle age')
plt.show()
```
