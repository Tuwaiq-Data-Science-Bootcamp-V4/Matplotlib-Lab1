```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
taif = [36,35,33,30,29,27,29]
riyad = [44,45,40,44,45,43,45]
d= [20,19,23,19,17,18,20]

dca = pd.DataFrame(data = {"Day":days,"Temperature":d})
dt = pd.DataFrame(data = {"Day":days,"Temperature":taif})
dr = pd.DataFrame(data = {"Day":days,"Temperature":riyad})
```

```python
# using list 
plt.plot(days,taif)
plt.title("Temps in Taif")
plt.xlabel("Days")
plt.ylabel("Temperature");
```

```python
#using DataFrame
plt.plot(dr["Day"],dr["Temperature"])
plt.title("Temps in Riyadh")
plt.xlabel("Days")
plt.ylabel("Temperature");
```

## Q2: Create pie chart with equal explode values.

```python
labels= ["A","B","C","D"]
b = [20,48,56,36]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.pie(b,labels=labels,autopct='%1.1f%%',  colors = colors , explode = (0.1, 0.1, 0.1, 0.1))
plt.title("")
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(dt["Day"],dt["Temperature"],marker = "o",alpha = 0.4,s=50)
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(dt["Day"],dt["Temperature"],marker = "o",alpha = 0.4,s=50)
plt.xlabel("Days",fontsize=16)
plt.ylabel("temp",fontsize=16)
plt.title("Temps in Taif",loc="left",color="red")
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
plt.plot(dr["Day"],dr["Temperature"])
plt.plot(dt["Day"],dt["Temperature"])
plt.plot(dca["Day"],dca["Temperature"])
plt.xlabel("Days",fontsize=16)
plt.ylabel("temp",fontsize=16)
plt.title("Temps in Taif ,Riyadh and DC",loc= "left")
plt.legend(["Riyadh","Taif","D"])
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=[20,10])




plt.subplot(1,3,2)
plt.pie(b,labels=labels,autopct='%1.1f%%',  colors = colors )



plt.subplot(1,3,1)
plt.plot(dr["Day"],dr["Temperature"])
plt.plot(dt["Day"],dt["Temperature"])

plt.xlabel("Days",fontsize=16)
plt.ylabel("temp",fontsize=16)
plt.title("Temps in Taif & Riyadh ",loc= "left")
plt.legend(["Riyadh","Taif","D"])


plt.subplot(1,3,3)

plt.scatter(dt["Day"],dt["Temperature"],marker = "o",alpha = 0.4,s=50)
plt.xlabel("Days",fontsize=16)
plt.ylabel("temp",fontsize=16)
plt.title("Temps in Taif",loc="left")
plt.show()
```
