```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
one = {'Name': ['Hussam', 'sam'],
        'Age': [23, 20]}
df = pd.DataFrame(one)

plt.bar(df['Name'], df['Age']);

plt.xlabel('Name')
plt.ylabel('Age')
plt.title('student age')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
data = {'Name': ['Ali', 'Abdullah', 'Sad', 'Saed'],
        'Age': [20, 20, 20, 20]}
df = pd.DataFrame(data)


plt.pie(df['Age'], labels=df['Name'], autopct='%1.1f%%',explode = [0.1,0.1,0.1,0.1])


plt.title('Same Age')
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = ['KSA', 'USA', 'UK', 'ARE']
y = [90, 20, 30, 80]

plt.scatter(x, y, marker='o', alpha=0.4, s=50)


plt.xlabel('Country')
plt.ylabel('Power')
plt.title('best and worst country with power')
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
x = ['KSA', 'USA', 'UK', 'ARE']
y = [90, 20, 30, 80]

plt.scatter(x, y, marker='o', alpha=0.4, s=50)


plt.xlabel('Country', fontsize=16)
plt.ylabel('Power', fontsize=16)
plt.title('best and worst country with power',color='red', loc='left')
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
x = ['KSA', 'USA', 'UK', 'ARE', 'India']
y1 = [10, 20, 30, 40, 50]
y2 = [15, 25, 35, 45, 55]
y3 = [5, 15, 25, 35, 45]

plt.plot(x, y1, label='L1', color='black')
plt.plot(x, y2, label='L2', color='g')
plt.plot(x, y3, label='L3', color='r')


plt.legend(loc='upper left')
plt.xlabel('Country')
plt.ylabel('Age')
plt.title('Avrage Age')
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=[20,10])

plt.subplot(2,3,1);
plt.plot(['KSA','USA','UK'],[1,2,3],color='red');
plt.legend(['power'],loc= 'upper left');

plt.subplot(2,3,2);
plt.plot(['KSA','USA','UK'],[3,2,1],color='blue');
plt.legend(['power'],loc= 'upper right');

plt.subplot(2,3,3);
plt.plot(['KSA','USA','UK'],[4,2,2],color='black');
plt.legend(['power'],loc= 'center left');
```
