```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
ages = [20,30,36,37,44,37,56,60,30,24,46,28,25,58,10,28,45,47,50,32,23,
        48,34,79,22,10,67,47,33]
```

```python
# Histogram using list
plt.figure(figsize=(5,4))
plt.hist(ages, bins = 8) 
plt.xlabel('ages')
plt.ylabel('count')
plt.title('Distribution of ages');
```

```python
# import the dataset
df = pd.read_csv('tmdb-movies (1).csv')
df.head(3)
```

```python
plt.figure(figsize=(5,4))
plt.hist(df["runtime"], bins = 10) 
plt.xlabel('runtime')
plt.ylabel('count')
plt.title('Distribution of runtime');
```

## Q2: Create pie chart with equal explode values.

```python
df.groupby('director').sum().sort_values(by='vote_average',ascending = False)['vote_average'][0:5]
```

```python
directors = ['Woody Allen','Clint Eastwood','Martin Scorsese','Steven Spielberg','Ridley Scott']
vote_average = [ 290.0 ,221.3,201.8,197.9,149.0]
```

```python
plt.pie(vote_average, labels= directors, autopct='%1.2f%%', startangle=90,
        radius=1, explode = [0,0,0,0,0]);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.figure(figsize=(5,4))
plt.scatter(df['budget'],df["runtime"], alpha =0.4 ,s =50)
plt.title('Budget vs.Runtime',fontsize = 10)
plt.xlabel('Budget',fontsize = 10)
plt.ylabel('Runtime',fontsize = 10);
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.figure(figsize=(5,4))
plt.scatter(df['budget'],df["runtime"], alpha =0.4 ,s =50)
plt.title('Budget vs.Runtime',loc = 'left',color = "red",fontsize = 20)
plt.xlabel('Budget',fontsize = 16)
plt.ylabel('Runtime',fontsize = 16);
```

## Q5: Create three line charts with Legends at left and colors.

```python
data={"INDICATIVE_PRICE":[800000,400000,20000,12000],
       "FINANCIAL_CEILING":[40000,500000,30000,12000],
        "LINK_PRICE":[40000,400000,20000,12000],
        "PLAN_YEAR":[2005,2010,2023,2024]}
df1= pd.DataFrame(data)
```

```python
df1
```

```python
plt.plot(df1['PLAN_YEAR'],df1['INDICATIVE_PRICE'] , color = 'red', linestyle = '--', linewidth = 2)
plt.plot(df1['PLAN_YEAR'],df1['FINANCIAL_CEILING'] , color = 'blue', linestyle = '--', linewidth = 2)
plt.plot(df1['PLAN_YEAR'],df1['LINK_PRICE'] , color = 'black', linestyle = '--', linewidth = 2)
plt.legend(['INDICATIVE_PRICE','FINANCIAL_CEILING','LINK_PRICE'], shadow = True, loc = 'upper left')
plt.xlabel('years')
plt.ylabel('amount')
plt.title('Then plans financial status every year')
plt.show()
```
